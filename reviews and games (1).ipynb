### Sentiment Assignment

```
!pip install vaderSentiment
!pip install nltk

import pandas as pd
import nltk
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 

nltk.downloader.download('vader_lexicon')
```

```
df_reviews['vader_score'] = [round(SentimentIntensityAnalyzer().polarity_scores(x)['compound'], 2) 
                             for x in df_reviews['clean_tokens']]

## values for sentiment score provided under the About the Scoring section of VADER documentation
## link here: https://github.com/cjhutto/vaderSentiment

def sentiment_tag(row):
    if row['vader_score'] >= 0.05:
        sentiment = "Positive"
    elif row['vader_score'] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    return sentiment

df_reviews['sentiment'] = df_reviews.apply(sentiment_tag, axis = 1)
```
